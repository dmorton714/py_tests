```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('get_data.csv')
df = df.sort_values('age')
```

you can make querys on defined values with the @

```python
min_age = 0
max_age = 99 
df = df.query('age > @min_age and age < @max_age')
df = df.sort_values('age')
df
```
